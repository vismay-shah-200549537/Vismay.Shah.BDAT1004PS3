```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Question 6

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
plt.figure(figsize=(10, 6))

plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per 1000')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per 1000')

plt.xlabel('Year')
plt.ylabel('Rate per 1000')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

plt.grid(True)
plt.show()
```

```python
# Question 7

selected_years = [1900, 1950, 2000]
selected_data = data[data['Year'].isin(selected_years)]

plt.figure(figsize=(10, 6))

bar_width = 3
plt.bar(selected_data['Year'] - bar_width/2, selected_data['Marriages_per_1000'], width=bar_width, color='blue', label='Marriages per 1000')
plt.bar(selected_data['Year'] + bar_width/2, selected_data['Divorces_per_1000'], width=bar_width, color='orange', label='Divorces per 1000')

plt.xlabel('Year')
plt.ylabel('Rate per 1000')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.legend()

plt.xticks(selected_data['Year'])
plt.grid(axis='y')

plt.show()
```
